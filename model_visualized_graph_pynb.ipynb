```python
import matplotlib.pyplot as plt
import networkx as nx

# Create the graph
G = nx.DiGraph()

# Add nodes and edges
G.add_node("Input Layer (28x28)", pos=(0, 2))
G.add_node("Quanv Layer\n(3 layers)", pos=(1, 2))
G.add_node("Flatten Layer", pos=(2, 2))
G.add_node("Fully Connected Layer\n(14*14*4 to 2)", pos=(3, 2))
G.add_node("Output Layer (2)", pos=(4, 2))

G.add_edges_from([
    ("Input Layer (28x28)", "Quanv Layer\n(3 layers)"),
    ("Quanv Layer\n(3 layers)", "Flatten Layer"),
    ("Flatten Layer", "Fully Connected Layer\n(14*14*4 to 2)"),
    ("Fully Connected Layer\n(14*14*4 to 2)", "Output Layer (2)")
])

# Extract node positions
pos = nx.get_node_attributes(G, 'pos')

# Draw the graph
plt.figure(figsize=(12, 6))
nx.draw(G, pos, with_labels=True, node_size=5000, node_color="skyblue", font_size=10, font_weight="bold", arrowsize=20)
plt.title("Visualization of the Quantum Hybrid Model")
plt.show()
```
